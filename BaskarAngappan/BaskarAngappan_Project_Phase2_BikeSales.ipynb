```python
# Assignment 1 - Bike sales prediction

# Lasso/Ridge/ElasticNet Regression seems to be best so far with higher accuracy, but it varies on each try
# Whereas Polynominal with little lesser accuracy, but works consistently

# Trial 1 - Linear Regression
import pandas as pd
from sklearn import preprocessing, model_selection as cross_validation, svm
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

parse_dates = ['date']
df = pd.read_csv('bikes.csv', parse_dates=parse_dates)
print(df.head())

#Plot the weather against bike count to see for relationship
#Plot the input data
plt.figure(figsize=(8,6))
plt.plot(df['temperature'], df['count'], 'o')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.show()

#df['date'] = pd.to_datetime(df.date)
#df['date'] = pd.to_datetime(df.date)
print(df.dtypes)
#df['date'] = pd.to_datetime(df['date'], yearfirst=True).dt.strftime('%Y-%m-%d')
#df['year'] = df.date.dt.year
#df['month'] = df.date.dt.month
#df['day'] = df.date.dt.day
print(df)
df.drop(['date'], 1, inplace=True)
x = np.array(df.drop(['count'], 1))
y = np.array(df['count'])
#print(x)
#print(y)
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x, y, test_size=0.2)
clf = LinearRegression(n_jobs=-1)
clf.fit(x_train, y_train)
accuracy = clf.score(x_test, y_test)
print("Accuracy using Linear Regression is : ", accuracy)
```

```python
# Trial 2 - Poly Linear Regression

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3)
x_ = poly.fit_transform(x)
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x_, y, test_size=0.2)
clf = LinearRegression(n_jobs=-1)
clf.fit(x_train, y_train)
accuracy = clf.score(x_test, y_test)
print("Accuracy using Poly Linear Regression is : ", accuracy)
```

```python
# Trial 3 - Using Ridge regression

from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline

clf = make_pipeline(PolynomialFeatures(degree=1), Ridge())
clf.fit(x_train, y_train)
accuracy = clf.score(x_test, y_test)
print("Accuracy using Ridge Linear Regression is : ", accuracy)
```

```python
# Trial 4 - Lasso Regression

from sklearn.linear_model import Lasso
clf = make_pipeline(PolynomialFeatures(degree=2), Lasso())
clf.fit(x_train, y_train)
accuracy = clf.score(x_test, y_test)
print("Accuracy using Lasso Linear Regression is : ", accuracy)
```

```python
# Trial 5 - ElasticNet Regression

from sklearn.linear_model import ElasticNet
clf = make_pipeline(PolynomialFeatures(degree=2), ElasticNet())
clf.fit(x_train, y_train)
accuracy = clf.score(x_test, y_test)
print("Accuracy using ElasticNet Linear Regression is : ", accuracy)
```

```python
# Trial 6 - Decision Tree Regression

clf = DecisionTreeRegressor()
clf.fit(x_train, y_train)
accuracy = clf.score(x_test, y_test)
print("Accuracy using Decision tree regressor is : ", accuracy)
```
